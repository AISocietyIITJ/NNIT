```python
import torch
import torch.nn as nn
import torch.nn.functional as F
class Neural_Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,32,5) #here 5 is the kernel size
        self.conv2 = nn.Conv2d(32,64,5)
        self.conv3 = nn.Conv2d(64,128,5)
        self._to_linear = None
     
        x = torch.randn(50,50).view(-1,1,50,50)
        self._to_linear = None
        self.convs(x)
        
        self.fc1 = nn.Linear(self._to_linear,512) #introduces linearity
        self.fc2 = nn.Linear(512,36)
        
    def convs(self, x):
        x = F.max_pool2d(F.relu(self.conv1(x)),(2,2))
        x = F.max_pool2d(F.relu(self.conv2(x)),(2,2))
        x = F.max_pool2d(F.relu(self.conv3(x)),(2,2))
     
        if self._to_linear is None:
            self._to_linear = np.prod(x[0].shape)
        
        return x
    
    def forward(self, x):
        x = self.convs(x)
        x = x.view(-1, self._to_linear) #reshapes X by flattening
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.softmax(x, dim = 1)

    
if torch.cuda.is_available():
    device = torch.device("cuda:0")
    print("running on GPU")
else:
    device = torch.device("cpu")
    print("running on CPU")
    
    
neuralnetwork = Neural_Network()
```

```python
import torch.optim as optim
import torch.nn as nn
from torch.nn import CrossEntropyLoss

X = torch.Tensor([i[0] for i in ocr1.training_data]).view(-1,50,50)
X = X/255.0
Y = torch.Tensor([i[1] for i in ocr1.training_data])

train_X = X
train_Y = Y
print(len(train_X))
```

```python
loss_function = nn.MSELoss() #computes loss

def train(neuralnetwork):
    BATCH_SIZE = 100
    EPOCHS = 10
    optimizer = optim.Adam(neuralnetwork.parameters(), lr=0.001)
    loss_function = nn.MSELoss()

    for epoch in range(EPOCHS):
        for i in tqdm(range(0, len(train_X), BATCH_SIZE)):
            batch_x = train_X[i:i + BATCH_SIZE].view(-1, 1, 50, 50)
            batch_y = train_Y[i:i + BATCH_SIZE]

            optimizer.zero_grad()
            outputs = neuralnetwork(batch_x)
            loss = loss_function(outputs, batch_y)
            loss.backward()
            optimizer.step() #updates

        print(f"epoch: {epoch}, loss: {loss}")

train(neuralnetwork)
        
```

```python
#testing_data = np.load("testing_data.npy", allow_pickle=True)

X = torch.Tensor([i[0] for i in ocr2.testing_data]).view(-1,50,50)
X = X/255.0
Y = torch.Tensor([i[1] for i in ocr2.testing_data])

test_X = X
test_Y = Y


def test(neuralnetwork):
    correct = 0
    total = 0
    with torch.no_grad():
        for i in tqdm(range(len(test_X))):
            label = torch.argmax(test_Y[i]).to(device)
            neuralnetwork_out = neuralnetwork(test_X[i].view(-1, 1, 50, 50).to(device))[0]

            pred_class = torch.argmax(neuralnetwork_out)
            if pred_class == label:
                correct += 1
            total += 1
    print("Accuracy:",(correct/total)*100)
```

```python
test(neuralnetwork)
```
